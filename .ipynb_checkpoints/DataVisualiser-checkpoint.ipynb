```python
#This takes all of the dss and profiles required to run a feeder analysis and downloads it
#You have to turn off your VPN for the web scraper to work

#If on local machine, comment out pass, if on HPC, include the pass, and run the Scraper.py file
#pass
"""
This program sorts through the AWS S3 Explorer for the Open Energy Data Initiative website to
"""

class dataSource:
    def __init__(self,name,link,profile):
        self.name = name
        self.link = link
        self.request = requests.get(self.link).text
        self.offset = 0
        self.profile = profile
        self.profileRequest = requests.get(f'{self.profile}{self.offset}').text
        
#Below are all of the datasets I used
set0 = dataSource('GSO_Rural_Base_rhs1_rdt137','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&prefix=SMART-DS%2Fv1.0%2F2018%2FGSO%2Frural%2Fscenarios%2Fbase_timeseries%2Fopendss%2Frhs1_1247%2Frhs1_1247--rdt137%2F','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&limit=100&prefix=SMART-DS%2Fv1.0%2F2018%2FGSO%2Frural%2Fprofiles%2F&offset=')
set1 = dataSource('GSO_Rural_Medium_sol_rhs0_rdt1528','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&prefix=SMART-DS%2Fv1.0%2F2018%2FGSO%2Frural%2Fscenarios%2Fsolar_medium_batteries_low_timeseries%2Fopendss%2Frhs0_1247%2Frhs0_1247--rdt1527%2F','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&limit=100&prefix=SMART-DS%2Fv1.0%2F2018%2FGSO%2Frural%2Fprofiles%2F&offset=')
set2 = dataSource('GSO_Industrial_Base_ihs0_idt629','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&prefix=SMART-DS%2Fv1.0%2F2018%2FGSO%2Findustrial%2Fscenarios%2Fbase_timeseries%2Fopendss%2Fihs0_1247%2Fihs0_1247--idt629%2F','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&limit=100&prefix=SMART-DS%2Fv1.0%2F2018%2FGSO%2Findustrial%2Fprofiles%2F&offset=')
set3 = dataSource('GSO_Industrial_Medium_sol_ihs3_idt226','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&prefix=SMART-DS%2Fv1.0%2F2018%2FGSO%2Findustrial%2Fscenarios%2Fsolar_medium_batteries_low_timeseries%2Fopendss%2Fihs3_1247%2Fihs3_1247--idt226%2F','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&limit=100&prefix=SMART-DS%2Fv1.0%2F2018%2FGSO%2Findustrial%2Fprofiles%2F&offset=')
set4 = dataSource('SFO_Urban_Base_p2uhs2_p2udt3761','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&prefix=SMART-DS%2Fv1.0%2F2018%2FSFO%2FP2U%2Fscenarios%2Fbase_timeseries%2Fopendss%2Fp2uhs2_1247%2Fp2uhs2_1247--p2udt3761%2F','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&limit=100&prefix=SMART-DS%2Fv1.0%2F2018%2FSFO%2FP2U%2Fprofiles%2F&offset=')
set5 = dataSource('SFO_Urban_Medium_sol_p2uhs12_udt4058','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&prefix=SMART-DS%2Fv1.0%2F2018%2FSFO%2FP2U%2Fscenarios%2Fsolar_medium_batteries_low_timeseries%2Fopendss%2Fp2uhs12_1247%2Fp2uhs12_1247--p2udt4058%2F','https://data.openei.org/s3_viewer?bucket=oedi-data-lake&limit=100&prefix=SMART-DS%2Fv1.0%2F2018%2FSFO%2FP2U%2Fprofiles%2F&offset=')
datasets = [set0,set1,set2,set3,set4,set5]
#To use, delete the comments, enter the name of the feeder, the link to the dss files, and the link to the profiles below as seen in the above examples (find links from https://data.openei.org/s3_viewer?bucket=oedi-data-lake&limit=100&prefix=SMART-DS%2F)
"""
NAME_OF_FEEDER = ''
DSS_FILE_LINK = ''
PROFILE_FILE_LINK = ''
set6= dataSource(f'{NAME_OF_FEEDER}',f'{DSS_FILE_LINK}',f'{PROFILE_FILE_LINK}&offset=')
datasets = [set0,set1,set2,set3,set4,set5,set6]
"""



#PUT THE DSS files into the DSSfiles folders

for sets in range (0,len(datasets),1):
    
    #Write the dss files from the bucket into the DSS folder
    os.chdir(f'{currentDirectory}\pydss-projects\{datasets[sets].name}\DSSfiles')
    soup = BeautifulSoup(datasets[sets].request,'lxml') #Request the lxml file from the link
    datalist = []
    for dataTable in soup.find_all('a', href=True):
        datalist.append(dataTable['href']) #list all of the downloadable files in datalist
    for x in range(0,len(datalist),1): 
        if '.dss' in datalist[x]: #only download the files with .dss at the end
            response = requests.get(datalist[x])
            filename = datalist[x].split('/')
            with open(filename[-1], "w") as file:
                file.write(response.text)
                
    #Check for every CSV file that the loadshapes.dss requires, and scrub out the exact file name required, place that in neededProfiles list
    loadShape = open('LoadShapes.dss','r+')
    neededProfiles,scrubList = [],[]
    while loadShape:
        wholeLine= loadShape.readline()
        scrub1 = wholeLine.split('/')
        for thing in range(len(scrub1)):
            scrubList.append(scrub1[thing].split(')'))
        for scrub2 in range(len(scrubList)):
            if len(scrubList[scrub2]) == 2 and scrubList[scrub2][0] not in neededProfiles:
                neededProfiles.append(scrubList[scrub2][0])
        if wholeLine == '': break
    loadShape.close()

    #Put csv files into profiles folder
    os.chdir(f'{currentDirectory}\profiles')
    profileSoup = BeautifulSoup(datasets[sets].profileRequest,'lxml')
    profileList = []
    csvList = []
    matches = 0
    profiles_not_found = []
    for item in range(0,len(neededProfiles),1):
        profiles_not_found.append(neededProfiles[item])
    status = profileSoup.find(role='status').text
    number_of_entries = status.split('of ')
    almost_fixed = number_of_entries[1].replace(' entries','')

    while matches < len(neededProfiles) and (datasets[sets].offset - 100) < int(almost_fixed.replace(',','')):
        #Create a list of all of the links to .csv files in csvList
        for profileTable in profileSoup.find_all('a', href = True):
            profileList.append(profileTable['href'])
        for csv in range(0,len(profileList),1):
            if '.csv' in profileList[csv]:
                csvList.append(profileList[csv])
    
        #Check if any of the csv files in neededProfiles are in csvList
        for y in range(0,len(csvList)):
            for bruh in range (0,len(neededProfiles),1):
                if neededProfiles[bruh] in csvList[y]:
                    #if they do match, open the link to the .csv file and write it into the profiles folder, increment matches
                    profiles_not_found.remove(neededProfiles[bruh])
                    if neededProfiles[bruh] not in os.listdir(f'{currentDirectory}\profiles'):
                        getProfile = requests.get(csvList[y])
                        profileFileName = csvList[y].split('/')
                        with open(profileFileName[-1],'w') as profileFile:
                            profileFile.write(getProfile.text)
                        break
                    matches += 1
        profileList.clear()
        csvList.clear()
        datasets[sets].offset += 100
        #go to new, offset page
        datasets[sets].profileRequest = requests.get(f'{datasets[sets].profile}{datasets[sets].offset}').text
        profileSoup = BeautifulSoup(f'{datasets[sets].profileRequest}','lxml')
                #if none of the neededProfiles are in this profile list, clear the profile list, go to the next page
        
    print(f'There were {len(profiles_not_found)} profiles not found for {datasets[sets].name}, they are:')
    for n in range(0,len(profiles_not_found),1):
        print(profiles_not_found[n])
    newLine = []
    with open(f'{currentDirectory}\pydss-projects\{datasets[sets].name}\DSSfiles\LoadShapes.dss','r') as profileFile:
        wholeLine=profileFile.readlines()
        for x in range(len(wholeLine)):
            newLine.append(wholeLine[x].replace('../../../../..',f'{currentDirectory}'))
    with open(f'{currentDirectory}\pydss-projects\{datasets[sets].name}\DSSfiles\LoadShapes.dss','w') as writeFile:
        for y in range(len(newLine)):
            writeFile.write(newLine[y])

print('Done!')
os.chdir(f'{holdDirectory}')
```
